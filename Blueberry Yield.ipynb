```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
X_train = train_df.drop(['id','yield'],axis=1)
y_train = train_df['yield']
```

```python
X_test = test_df.drop('id',axis=1)
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 
sns.heatmap(X_train.corr(),annot=True)
```

```python
corr_matrix = X_train.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```python
to_drop
```

```python
X_train.drop(to_drop, axis=1, inplace=True)
X_test.drop(to_drop, axis=1, inplace=True)
```

```python
sns.boxplot(data=y_train,orient='h')
```

```python
sns.kdeplot(data=y_train)
```

```python
y_train.skew(), y_train.kurt()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg_model = reg.fit(X_train,y_train)
```

```python
reg_model.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
np.mean(cross_val_score(reg,X_train,y_train))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(y_train,reg_model.predict(X_train))
```

```python
mean_absolute_error(y_train,reg_model.predict(X_train))
```

```python
reg_model_pred = reg_model.predict(X_test)
```

```python
submission  = pd.read_csv('sample_submission.csv')
```

```python
submission['yield'] = reg_model_pred
```

```python
submission.to_csv('MLR.csv',index=False)
```
